# Les branches

- Principes
- Commandes
- Fucionner

Une branche est bifurcation dans la lignée de commits afin de :

- Garder une branche `master` stable et propre et créer des branches pour les développements de fonctionnalités
- Tester une idée avant de l'implémenter définitivement
- Avoir des branches par développeur et une branche commune
- Maintenir des anciennes versions de code en même temps que des nouvelles



- Une branche est un pointeur vers le dernier commit réalisé dans une bifurcation donnée
- Les branches sont à considérer comme des univers parallèles, __indépendants__ les uns des autres
- On peut les créer, les renommer, les supprimer, les fusionner, les déplacer

- `git branch` : liste toutes les branches locales du projet
- `git branch <new_branch>` : crée une nouvelle branche
- `git switch <une_branche>` : change de branche
- `git diff <branch1> <branch2>` : différences entre les fichiers de 2 branches
- `git merge <une_branche>` : fusionne la branche active avec `<une_branche>`
- `git merge <branch1> <branch2>` : fusionne les deux branches
- `git branch -M <new_name>` : renomme la branche active
- `git branch -d <branch>` : efface une branche
