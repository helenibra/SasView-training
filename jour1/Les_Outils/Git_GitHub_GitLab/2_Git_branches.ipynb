# Les branches

- Principes
- Commandes
- Fucionner

Une branche est bifurcation dans la ligne de commits afin de :

- Garder une branche `master` stable et propre et crer des branches pour les dveloppements de fonctionnalits
- Tester une ide avant de l'implmenter dfinitivement
- Avoir des branches par dveloppeur et une branche commune
- Maintenir des anciennes versions de code en mme temps que des nouvelles



- Une branche est un pointeur vers le dernier commit ralis dans une bifurcation donne
- Les branches sont  considrer comme des univers parallles, __indpendants__ les uns des autres
- On peut les crer, les renommer, les supprimer, les fusionner, les dplacer

- `git branch` : liste toutes les branches locales du projet
- `git branch <new_branch>` : cre une nouvelle branche
- `git switch <une_branche>` : change de branche
- `git diff <branch1> <branch2>` : diffrences entre les fichiers de 2 branches
- `git merge <une_branche>` : fusionne la branche active avec `<une_branche>`
- `git merge <branch1> <branch2>` : fusionne les deux branches
- `git branch -M <new_name>` : renomme la branche active
- `git branch -d <branch>` : efface une branche
