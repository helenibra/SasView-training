## Principes - Commit



- Nous écrivons un fichier source `<file>` (testé, modifié, testé, modifié...)
- Quand nous sommes satisfait du code, nous __validons les modifications__ réalisées.
  - nous "copions" le fichier dans un répertoire virtuel appelé _Staging Area_
  - chaque fichier de ce répertoire est __tracké__ pour toute la suite du projet
  - Nous pouvons reprendre l'écriture de `<file>`
- Quand nous sommes prêt, nous __versionnons__ le stage en créant un __commit__



- __Un commit = "différence de données" + id + commentaire + date + nom + email + tous les pointeurs qui y réfèrent...__
- Un commit n'a aucune incidence sur le stage ou le working tree

- __Nous ne pouvons pas supprimer un commit__

## Principes - Commandes

- Pour connaître le statut de l'état de versionning : `git status`

- Pour valider les modifications dans le stage : `git add`

- Pour afficher le contenu du stage : `git ls-files`

- Pour changer un chemin dans un stage : `git mv <path1> <path2>`

- Pour supprimer un chemin dans un stage : `git rm <path>`

__Versionner les modifications dans un commit__ :

- `git commit -m "commentaire"` : toujours ajouter un commentaire à un commit
- `git commit -am "commentaire"` : ajoute au stage tous les fichiers modifiés déjà suivis et commit avec un commentaire
- `git commit --amend [-m "commentaire"]` : modification du commentaire ou du contenu du dernier commit par la création d'un nouveau commit

__Historique des commits__ :

- `git log` : historique des commmits

- `git show [<commit>]` : affiche les changements apportés au commit indiqué (HEAD par défaut)

- `git reflog` : historique des HEAD pour ne pas perdre les références aux commits

__Les différences__

- `git diff` : différence entre le working tree et le stage
- `git diff --cached` : différence entre le stage et le dernier commit
- `git diff <commit>` : différence entre le working tree et `<commit>`
- `git diff <commit> <file>` : différence entre le `<file>` du working tree et celui de `<commit>`
- `git diff <commit1> <commit2>` : différence entre les éléments de `<commit1>` et `<commit2>`

__Restauration__

- `git restore <file>` : restaurer `<file>` depuis le stage
- `git restore --source=master` restaurer tout le répertoire local depuis master

__Ignorer des fichiers automatiquement au staging__

Il peut être intéressant d'ignorer automatiquement une liste de fichier (sur un `git add .` par exemple). Pour cela il faut créer un fichier caché `.gitignore` à la racine du projet :

- `*.a` : ignore tous les fichiers d'extensions `a`
- `!lib.a` : sauf le fichier `lib.a`
- `TODO/` : ignore tous les fichiers du répertoire `TODO` 
- `doc/*.txt` : ignore tous les `*.txt`du répertoire `doc/` mais pas `doc/server/arch.txt` par exemple
- `doc/**/*.pdf` ignore les fichiers `*.pdf` récursivement dans le répertoire `doc/`

__Les tags__

- Il est souvent utile de "nommer", numéroter ou étiqueter un commit (tag)
- Nous pourrons alors identifier le commit par son tag au lieu de son id !
- Un tag est un pointeur.
- `git tag <tag> <commit>` : créer un tag
  - `<tag>=1.1.0, stable...`
  - `<commit>=de8b923, master, HEAD^`
- `git tag` : liste des tags
- `git tag -d <tag>` : supprime un tag

__les numéros de versions__

`2.4.3  => MAJEUR . MINEUR . PATCH [-NOM_DE_BRANCHE]`

- MAJEUR : version majeure
- MINEUR : ajout de fonctionnalités
- PATCH : correctif
- NOM_DE_BRANCHE : "dev", "unstable",...
