## Principes - Commit



- Nous crivons un fichier source `<file>` (test, modifi, test, modifi...)
- Quand nous sommes satisfait du code, nous __validons les modifications__ ralises.
  - nous "copions" le fichier dans un rpertoire virtuel appel _Staging Area_
  - chaque fichier de ce rpertoire est __track__ pour toute la suite du projet
  - Nous pouvons reprendre l'criture de `<file>`
- Quand nous sommes prt, nous __versionnons__ le stage en crant un __commit__



- __Un commit = "diffrence de donnes" + id + commentaire + date + nom + email + tous les pointeurs qui y rfrent...__
- Un commit n'a aucune incidence sur le stage ou le working tree

- __Nous ne pouvons pas supprimer un commit__

## Principes - Commandes

- Pour connatre le statut de l'tat de versionning : `git status`

- Pour valider les modifications dans le stage : `git add`

- Pour afficher le contenu du stage : `git ls-files`

- Pour changer un chemin dans un stage : `git mv <path1> <path2>`

- Pour supprimer un chemin dans un stage : `git rm <path>`

__Versionner les modifications dans un commit__ :

- `git commit -m "commentaire"` : toujours ajouter un commentaire  un commit
- `git commit -am "commentaire"` : ajoute au stage tous les fichiers modifis dj suivis et commit avec un commentaire
- `git commit --amend [-m "commentaire"]` : modification du commentaire ou du contenu du dernier commit par la cration d'un nouveau commit

__Historique des commits__ :

- `git log` : historique des commmits

- `git show [<commit>]` : affiche les changements apports au commit indiqu (HEAD par dfaut)

- `git reflog` : historique des HEAD pour ne pas perdre les rfrences aux commits

__Les diffrences__

- `git diff` : diffrence entre le working tree et le stage
- `git diff --cached` : diffrence entre le stage et le dernier commit
- `git diff <commit>` : diffrence entre le working tree et `<commit>`
- `git diff <commit> <file>` : diffrence entre le `<file>` du working tree et celui de `<commit>`
- `git diff <commit1> <commit2>` : diffrence entre les lments de `<commit1>` et `<commit2>`

__Restauration__

- `git restore <file>` : restaurer `<file>` depuis le stage
- `git restore --source=master` restaurer tout le rpertoire local depuis master

__Ignorer des fichiers automatiquement au staging__

Il peut tre intressant d'ignorer automatiquement une liste de fichier (sur un `git add .` par exemple). Pour cela il faut crer un fichier cach `.gitignore`  la racine du projet :

- `*.a` : ignore tous les fichiers d'extensions `a`
- `!lib.a` : sauf le fichier `lib.a`
- `TODO/` : ignore tous les fichiers du rpertoire `TODO` 
- `doc/*.txt` : ignore tous les `*.txt`du rpertoire `doc/` mais pas `doc/server/arch.txt` par exemple
- `doc/**/*.pdf` ignore les fichiers `*.pdf` rcursivement dans le rpertoire `doc/`

__Les tags__

- Il est souvent utile de "nommer", numroter ou tiqueter un commit (tag)
- Nous pourrons alors identifier le commit par son tag au lieu de son id !
- Un tag est un pointeur.
- `git tag <tag> <commit>` : crer un tag
  - `<tag>=1.1.0, stable...`
  - `<commit>=de8b923, master, HEAD^`
- `git tag` : liste des tags
- `git tag -d <tag>` : supprime un tag

__les numros de versions__

`2.4.3  => MAJEUR . MINEUR . PATCH [-NOM_DE_BRANCHE]`

- MAJEUR : version majeure
- MINEUR : ajout de fonctionnalits
- PATCH : correctif
- NOM_DE_BRANCHE : "dev", "unstable",...
