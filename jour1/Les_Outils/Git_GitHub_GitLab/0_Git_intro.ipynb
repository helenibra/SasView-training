- que nous soyons une équipe de développeurs créatifs

Nécessité : trouver un outil pour gagner en productivité

- __qui facilite le travail collaboratif__

### Présentation de Git

- Problématique : Développement du noyau Linux
- Ancré sur de l'existant 
    - 1er dev=1972
    - RCS=1982
    - CVS=1990
    - BitKeeper=1998
    - SVN=2000
    - Mercurial=2005
- Développement :
   - Avril 2005 : Besoin pressant d'un outil
   - La base a été codée en 15 jours par Linus Torvald
   - Juillet 2005 : projet confié à la communauté
   - Décembre 2005 : première version V1

- Caractéristiques : local + décentralisé, non linéaire, rapidité, adapté aux gros projets...
- popularité mondiale immédiate
- open-source
- multi-OS
- support local + https + ssh
- Facile à apprendre
- Système de gestion de versions le plus utilisé
- difficile à maîtriser

### Git différent de GitLab/GitHub



- __Git__ : logiciel de versionnement de fichiers
- __GitLab/GitHub__ : Forges

### Git différent de GitLab/GitHub

- plateformes de développement multi-projets (comme Redmine, TuLeap...)
- basées sur _Git_ pour le versionnement de fichiers
- gestion des utilisateurs, différents accès...
- suivi de projet, demandes de fonctionnalités, suivi de bugs, boîtes à idées...
- compilations, gestion de cas tests...
- visualisation du code, historique, édition en ligne
- __GitLab est un projet libre__
- GitHub appartient à Microsoft (__attention__)

   - `git help <command>` : aide sur une commande spécifique
   - `git help config`
   - `git help commit`

### Configuration

Les fichiers de configuration : 
  - La configuration __système__ se trouve dans `/etc/config`
  - La configuration __globale__ se trouve dans `~/.gitconfig`
  - la configuration d'un __projet__ se trouve à la racine du projet dans `.git/config`

Les champs à définir :
   - `git config --global user.name "Prenom NOM"`
   - `git config --global user.email "adresse_mail@exemple.fr"`
   - `git config --global core.editor vi`
   - `git config --global color.ui auto (activé par défaut)`
   - ...

Listes des paramètres définis :
   - `git config --list`

### Démarrer un dépôt Git 

Initialiser un projet local
   - `git init [options] <project>`

Cloner un projet en local depuis un dépôt distant
   - `git clone <server> [<myproject>]`
   - `git clone https://github.com/<user>/<project>`
   - `git clone https://github.com/<user>/<project>`
   - `git clone [ssh://]<user>@<host>[:<path>/<project>]`
   - `git clone git@https://github.com:<user>/<project>`
   - `git clone /tmp/<project>`

=> Création d'un répertoire `<projet>/.git`
